# Base Modeller

#### Veri Hazƒ±rlama (Data Preprocessing)

```python
df["publishedAt"] = pd.to_datetime(df["publishedAt"], errors="coerce")

df["year"] = df["publishedAt"].dt.year
df["month"] = df["publishedAt"].dt.month
df["day"] = df["publishedAt"].dt.day
df["hour"] = df["publishedAt"].dt.hour
df["is_weekend"] = df["publishedAt"].dt.weekday.isin([5, 6]).astype(int)

df = df[df["is_trending"].isin([0, 1])]  # sa√ßma deƒüer varsa temizle

print("Yƒ±l daƒüƒ±lƒ±mƒ±:")
print(df["year"].value_counts().sort_index())
```

&nbsp;
#### Train‚ÄìValidation‚ÄìTest Split (Zamana Dayalƒ± Ayrƒ±≈ütƒ±rma)

```python
train = df[df["year"] <= 2023].copy()
val   = df[df["year"] == 2024].copy()
test  = df[df["year"] == 2025].copy()

print("Train:", train.shape)
print("Val  :", val.shape)
print("Test :", test.shape)
```

&nbsp;
#### Feature Engineering (√ñzellik √áƒ±karma)

```python
train["title_length"] = train["title"].astype(str).apply(len)
val["title_length"]   = val["title"].astype(str).apply(len)
test["title_length"]  = test["title"].astype(str).apply(len)

train["description_length"] = train["description"].astype(str).apply(len)
val["description_length"]   = val["description"].astype(str).apply(len)
test["description_length"]  = test["description"].astype(str).apply(len)

def count_tags(x):
    x = str(x)
    if x == "[]" or x.strip() == "":
        return 0
    return len(x.split("|"))

train["tag_count"] = train["tags"].apply(count_tags)
val["tag_count"]   = val["tags"].apply(count_tags)
test["tag_count"]  = test["tags"].apply(count_tags)

for df_part in [train, val, test]:
    df_part["likes_ratio"]    = df_part["likes"] / (df_part["view_count"] + 1)
    df_part["comments_ratio"] = df_part["comment_count"] / (df_part["view_count"] + 1)
    df_part["engagement"]     = (df_part["likes"] + df_part["comment_count"]) / (df_part["view_count"] + 1)

cat_le = LabelEncoder()
full_cat = pd.concat([train["categoryId"], val["categoryId"], test["categoryId"]])
cat_le.fit(full_cat.astype(str))

train["categoryId_enc"] = cat_le.transform(train["categoryId"].astype(str))
val["categoryId_enc"]   = cat_le.transform(val["categoryId"].astype(str))
test["categoryId_enc"]  = cat_le.transform(test["categoryId"].astype(str))
```

&nbsp;
#### Feature Matrix Preparation (√ñzellik Matrisi Hazƒ±rlama)

```python
feature_cols = [
    "view_count", "likes", "comment_count",
    "year", "month", "day", "hour", "is_weekend",
    "title_length", "description_length", "tag_count",
    "likes_ratio", "comments_ratio", "engagement",
    "categoryId_enc"
]

target_col = "is_trending"

X_train, y_train = train[feature_cols].fillna(0), train[target_col]
X_val,   y_val   = val[feature_cols].fillna(0),   val[target_col]
X_test,  y_test  = test[feature_cols].fillna(0),  test[target_col]

print("Feature sayƒ±sƒ±:", len(feature_cols))
print("Train shape:", X_train.shape)
print("Val shape  :", X_val.shape)
print("Test shape :", X_test.shape)
```

&nbsp;
### Model 1 - Logistic Regression (Validation)

```python
logreg = LogisticRegression(max_iter=200, n_jobs=-1)
logreg.fit(X_train, y_train)

val_pred_lr = logreg.predict(X_val)

print("LOGISTIC REGRESSION (Validation)")
print(classification_report(y_val, val_pred_lr))

train_pred_lr = logreg.predict(X_train)
print("\nLOGISTIC REGRESSION (Train)")
print(classification_report(y_train, train_pred_lr))
```

&nbsp;
### Model 1 - Logistic Regression (Test)

```python
X_full = pd.concat([X_train, X_val], axis=0)
y_full = pd.concat([y_train, y_val], axis=0)

logreg_final = LogisticRegression(max_iter=200, n_jobs=-1)
logreg_final.fit(X_full, y_full)

test_pred_lr = logreg_final.predict(X_test)

print("LOGISTIC REGRESSION (Test)")
print(classification_report(y_test, test_pred_lr))
```

&nbsp;
### Model 2 - CatBoost Classifier (Validation)

```python
cat = CatBoostClassifier(
    depth=6,
    learning_rate=0.1,
    loss_function="Logloss",
    eval_metric="AUC",
    iterations=300,
    random_state=42,
    verbose=50
)

cat.fit(
    X_train, y_train,
    eval_set=(X_val, y_val),
    use_best_model=True
)

val_pred_cat = cat.predict(X_val)

print("\nCATBOOST (Validation)")
print(classification_report(y_val, val_pred_cat))

train_pred_cat = cat.predict(X_train)
print("\nCATBOOST (Train)")
print(classification_report(y_train, train_pred_cat))
```

&nbsp;
### Model 2 - CatBoost Classifier (Test)

```python
cat_final = CatBoostClassifier(
    depth=6,
    learning_rate=0.1,
    loss_function="Logloss",
    eval_metric="AUC",
    iterations=cat.tree_count_,
    random_state=42,
    verbose=False
)

cat_final.fit(X_full, y_full)

test_pred_cat = cat_final.predict(X_test)

print("\nCATBOOST (Test)")
print(classification_report(y_test, test_pred_cat))
```

&nbsp;
#### Confusion Matrix ve ROC‚ÄìAUC

```python
fig, axes = plt.subplots(1, 2, figsize=(11,4))

cm_lr = confusion_matrix(y_test, test_pred_lr)
disp_lr = ConfusionMatrixDisplay(cm_lr, display_labels=[0,1])
disp_lr.plot(ax=axes[0], cmap="Blues", colorbar=False)
axes[0].set_title("Logistic Regression - Test")

cm_cat = confusion_matrix(y_test, test_pred_cat)
disp_cat = ConfusionMatrixDisplay(cm_cat, display_labels=[0,1])
disp_cat.plot(ax=axes[1], cmap="Greens", colorbar=False)
axes[1].set_title("CatBoost - Test")

plt.tight_layout()
plt.show()

lr_proba = logreg_final.predict_proba(X_test)[:, 1]
cat_proba = cat_final.predict_proba(X_test)[:, 1]

fpr_lr, tpr_lr, _ = roc_curve(y_test, lr_proba)
fpr_cat, tpr_cat, _ = roc_curve(y_test, cat_proba)

auc_lr = auc(fpr_lr, tpr_lr)
auc_cat = auc(fpr_cat, tpr_cat)

plt.figure(figsize=(7,6))
plt.plot(fpr_lr, tpr_lr, label=f"LogReg (AUC = {auc_lr:.3f})")
plt.plot(fpr_cat, tpr_cat, label=f"CatBoost (AUC = {auc_cat:.3f})")
plt.plot([0,1], [0,1], "--", color="gray")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Test Set")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

#### Prediction Function (Yeni Video Tahmin Fonksiyonu - Catboost)

```python
def build_features_from_raw(
    view_count, likes, comment_count,
    year, month, day, hour, is_weekend,
    title, description, tags, category_id
):
    title_length = len(str(title))
    description_length = len(str(description))

    if tags is None or str(tags).strip() == "" or str(tags) == "[]":
        tag_count = 0
    else:
        tag_count = len(str(tags).split("|"))

    likes_ratio = likes / (view_count + 1)
    comments_ratio = comment_count / (view_count + 1)
    engagement = (likes + comment_count) / (view_count + 1)

    # unseen category fix
    cat_value = str(category_id)
    if cat_value in cat_le.classes_:
        category_enc = cat_le.transform([cat_value])[0]
    else:
        category_enc = -1

    row = pd.DataFrame([{
        "view_count": view_count,
        "likes": likes,
        "comment_count": comment_count,
        "year": year,
        "month": month,
        "day": day,
        "hour": hour,
        "is_weekend": is_weekend,
        "title_length": title_length,
        "description_length": description_length,
        "tag_count": tag_count,
        "likes_ratio": likes_ratio,
        "comments_ratio": comments_ratio,
        "engagement": engagement,
        "categoryId_enc": category_enc
    }])

    return row[feature_cols]

def predict_video_trend(model, **kwargs):
    X_new = build_features_from_raw(**kwargs)
    proba = model.predict_proba(X_new)[:, 1][0]
    pred = int(proba >= 0.5)
    print(f"Tahmin: {pred}")
    print(f"Trend olma olasƒ±lƒ±ƒüƒ±: %{proba*100:.2f}")
    return pred, proba
```
